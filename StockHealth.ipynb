```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import MinMaxScaler, StandardScaler

from sklearn.metrics import classification_report, confusion_matrix

import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Stocks Health.xlsx')
```

```python
df.head()
```

```python
features = ['Market Cap',
             'ROE %',
             'Gross Margin %',
             'Operating Margin %',
             'Net Profit Margin %',
             'PE',
             'PEG',
             'PS',
             'CR',
             'LT Debt ($ M)',
             'Free Cash Flow ($ M)'
            ]
```

```python
X = df[features]
y = df["Company Score"]

scaler = StandardScaler()
sX = scaler.fit_transform(X)
```

```python
le = LabelEncoder() #Encoude the "Good, Fair outputs as numbers"
y_encoded = le.fit_transform(y)
X_train, X_test, y_train, y_test = train_test_split(sX, y_encoded, test_size=10, random_state=42)
```

```python
print (y_test)
```

```python
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print("Accuracy:", accuracy)
# check results
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred)) 
```

```python
# SAVING PART
df.to_excel("Stocks_Health_Clustered.xlsx")

import pickle
pfile = "stockHealth_preTrainedModel.pkl"  

with open(pfile, 'wb') as file:  
    pickle.dump(model, file)
```
